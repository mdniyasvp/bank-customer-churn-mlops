## Categorical Feature Analysis Summary

### Geography
- Germany has the highest churn rate (~38%), while France and Spain have much lower churn (~17%).
- Geography appears to be a strong predictor of customer churn.

### Gender
- Female customers exhibit a higher churn rate (~28%) compared to male customers (~16%).
- Gender may contribute useful predictive information.

### HasCrCard
- Customers with and without credit cards have similar churn rates.
- Credit card ownership alone does not appear to strongly influence churn.

### IsActiveMember
- Inactive customers churn at a much higher rate (~30%) than active customers (~13%).
- Customer activity is one of the strongest indicators of retention.

### NumOfProducts
- Customers with two products have the lowest churn rate.
- Customers with three or four products show very high churn rates.
- This feature warrants further investigation because the higher product-count groups may contain relatively few customers.

# EDA Summary

## Dataset Quality
- No missing values.
- No duplicate records.
- Large dataset with 165,034 customer records.

## Target Distribution
- Customer churn rate is approximately 21%.
- The dataset is moderately imbalanced.
- Accuracy alone is not an appropriate evaluation metric.

## Numerical Features
- CreditScore is approximately normally distributed.
- Age is right-skewed with valid business outliers.
- Balance contains a large number of zero values, indicating two distinct customer groups.
- EstimatedSalary is nearly uniformly distributed.
- NumOfProducts is a discrete feature.

## Categorical Features
- Germany has the highest customer churn rate.
- Female customers churn more than male customers.
- Active members are much less likely to churn.
- Customers with two products have the lowest churn rate.
- Customers with three or four products exhibit very high churn rates.

## Correlation Analysis
- Age has the strongest positive correlation with churn.
- IsActiveMember and NumOfProducts show the strongest negative correlations.
- Correlation was used only as a supporting analysis because many machine learning models capture nonlinear relationships.
